```python

"""
    DocString:
    
    This game is based on the movie Squid game. 
    The goal of the game is to win the grand prize of $45,000. 
    Player earns $15,000 for each level won.
    
    Each level is defined as a function.
    
    1. doors_chamber: Player is required to guess right door where money bag is
    2. glass_bridge: Player needs to choose correct tile to be able to cross bridge
    3. switch_room: Player lures room custodian to sleep in order to access
                    the money bag.
    
    Other functions in the game are:
    play_again: this function brings the player back to the beginning of the game
    fail: player loses and exits game
  
    
"""
import random #imports random package

# defining the doors_chamber function, the first level of game
# player_name allows player to input their name, it is dynamic
def doors_chamber():
    print(    
    """
    Welcome to the Squid game! 
    The squid game has 3 challenge levels, and each level is worth $15,000.
    To win, you must pass all levels to win a grand prize of $45,000!!!
    Yes!!! the winner goes home with a whopping sum of $45,000 to spend this holiday!!!
    Just like that, right? Sounds quite easy!!!! 
    What's your name? """)
    
    # prompting for player input
    player_name = input(prompt = "Type your name\n")
    
    # dynamically applying player input
    print(f"""
    Welcome {player_name.capitalize()} to the first room, the chamber of doors
    Ready to make your first $15,000?
    """)
    # player is prompted to press "Enter"
    input(prompt = "Press enter to continue\n")
    
    # Player is introduced to first challenge 
    options = ['1) circle',
                '2) triangle',
                '3) square']
    print(f"""
    In this chamber, there are 3 doors represented by 3 shapes. 
    Only 1 of the doors leads to a room with a money bag of $15,000. 
    Find the door, and the money is all yours.
    Are you ready, which door looks rich?
    {options[0]}
    {options[1]}
    {options[2]}
        """)
    
    # declaring objects
    number = 3 #number remains the same for each guess in this level
    tries = 2 #player is allowed 2 tries

    # looping until a user is out of guesses

    while tries > 0: # repeatedly executes a target statement as long as a given condition is true
        guess = input(prompt = '\n Choose a number from 1 - 3 (Example: 2)\n> ')
        guess = guess.lower()
        tries -= 1 # reduces the number of tries by 1 after a wrong guess

        # conditional statement based on user input
        if guess in options[2]: 
            print("""\nYay, correct answer! You just won $15,000 fair and square
            Off you go to the glass bridge
            """)
            input(prompt = "Press enter to continue\n")
            glass_bridge()
            break # breaks the loop so player goes to next level
            
        else:

            if tries > 0:
                print(f"""\nIncorrect answer, please choose another number from 1 - 3 (Example: 2). 
                      You have {tries} try remaining.""") # prompts player if guess is number that is not 1 - 3

            elif tries == 0:
                print(f"\nYou are out of tries. The correct answer is 3(square).") #player is out of tries, prompt informs player of right answer
                play_again() # brings player back to the beginning of game
                break # breaks the loop so player doesn't keep playing again

            
# defining the glass_bridge function, the second level of game 
# player is introduced to glass_bridge
def glass_bridge():
    print("""
Welcome champ, that wasn't so difficult, right?
You are about to get on the glass bridge, the bridge has 3 coloured tiles,
but only 1 tile is strong enough to hold you without breaking.
One wrong move, you crash and the cash disappears.
If you cross successfully, you will win another $15,000. 
Are you ready? Which of the tiles looks like money?
Choose a number from 1 - 3

""")
    # declaring objects
    number = random.randint(1,3) #generates a random number between 1 and 3, player has to guess number right
    tries = 2 #player is allowed 2 tries

    # looping until a user is out of guesses

    while tries > 0: # repeatedly executes a target statement as long as a given condition is true
        guess = input(prompt = '\n Choose a number from 1 - 3 (Example: 2)\n> ')
        tries -= 1 # reduces the number of tries by 1 after a wrong guess
        
        
        try: # try-except block handle exception to prevent break in code
            guess = int(guess) # converting choice for numerical comparison
            
            # conditional statement based on user input
            if guess == number: 
                print("""\nYay, you made the right choice. 
                  $15,000 + $15,000. You are now $30,000 richerrrrr.
                  Off you go to the switch room!
                """)
                input(prompt = "Press enter to continue\n")
                switch_room()
                break # breaks the loop so player goes to next level

            else: # if player guesses number that is not 1 - 3
               
                if tries > 0:
                    print(f"""\nIncorrect answer, please choose another number from 1 - 3 (Example: 2). 
                          You have {tries} try remaining.""") # prompts player if guess is number that is not 1 - 3

                elif tries == 0:
                    print(f"\nYou are out of tries. The correct answer is {number}.") #player is out of tries, prompt informs player of right answer
                    play_again() # brings player back to the beginning of game
                    break # breaks the loop so player doesn't keep playing again
        
            
        except: # try-except block handle exception to prevent break in code
            print(f"""Invalid entry. Please try again. 
                  You have {tries} try remaining.\n""")
            
            if tries == 0:
                print(f"\nYou are out of tries. The correct answer is {number}).") #player is out of tries, prompt informs player of right answer
                play_again() # brings player back to the beginning of game
                break # breaks the loop so player doesn't keep playing again

                
# defining the switch_room function, the third and final level of game
# introducing player to switch_room                
def switch_room():
    print("""
    Welcome to the switch room! You are at the finish line, real close.
    The custodian of the final money bag lives here, and he is wary of strangers and guards the bag intently. 
    One of the switches plays a song that puts the custodian to sleep, 
    the others are alarm sounds that announce your arrival in the room. 
    Your challenge is to lure the custodian to sleep, while you grab the money bag.
    Guess the right song-switch, and you are all set for your vacation. 
    Time to face the music! 
    Choose a number 1 - 3

    """)
    
    # declaring objects
    number = random.randint(1,3) #generates a random number between 1 and 3, player has to guess number right
    tries = 2 #player is allowed 2 tries

    # looping until a user is out of guesses

    while tries > 0: # repeatedly executes a target statement as long as a given condition is true
        guess = input(prompt = '\n Choose a number from 1 - 3 (Example: 2)\n> ')
        tries -= 1 # reduces the number of tries by 1 after a wrong guess
        
        
        try: # try-except block handle exception to prevent break in code
            guess = int(guess) # converting choice for numerical comparison
            
            # conditional statement based on user input
            if guess == number:                 
                print("""\nCongratulations!!! You are a champion!!!!. 
                You're off to the start of a great vacation with your $45,000!!!
                """)
                input(prompt = "Press enter to exit game\n")
                break # breaks the loop so player goes to next level

            else: # if player guesses number that is not 1 - 3
               
                if tries > 0:
                    print(f"""\nIncorrect answer, please choose another number from 1 - 3 (Example: 2). 
                          You have {tries} try remaining.""") 

                elif tries == 0:
                    print(f"\nYou are out of tries. The correct answer is {number}.") #player is out of tries, prompt informs player of right answer
                    play_again() # brings player back to the beginning of game
                    break # breaks the loop so player doesn't keep playing again
        
            
        except: # try-except block handle exception to prevent break in code
            print(f"""Invalid entry. Please try again. 
                  You have {tries} try remaining.\n""") 
            
            if tries == 0:
                print(f"\nYou are out of tries. The correct answer is {number}).") #player is out of tries, prompt informs player of right answer
                play_again() # brings player back to the beginning of game
                break # breaks the loop so player doesn't keep playing again
                
# defining the play_again function
# brings player back to the beginning of game
def play_again():
    options = ['1) yes',
                '2) no']
    print(f"""
Do you want to play again?
{options[0]}
{options[1]}
    \n""")
    
    choice = input(prompt="Choose 1 or 2. (Example: 1)")
    
    # converting choice for numerical comparison
    choice = choice.lower()
    
    if choice in options[0]:
        print("\nYou are welcome back!\n")
        input(prompt = "<Press enter to continue>\n")
        doors_chamber() # This moves us back to the beginning of the game
        
    
    else:
        fail() # player exits game
    
#defining the fail function
#player loses and exits game
def fail():
    print('\nSorry!!!You have failed the squid game challenge.')
    input(prompt = 'Press enter to exit game')
       
        
               
doors_chamber()
```
